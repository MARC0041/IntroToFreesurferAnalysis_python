```python
from scipy import stats
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt
```

```python
# Analyse left hemisphere
# Create subplots for each column
num_columns = df_healthy_lh.shape[1]

# Create a figure with subplots
num_rows = 8
num_cols = 5
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 18))

# Iterate through columns
for i in range(num_rows):
    for j in range(num_cols):
        col_index = i * num_cols + j
        if col_index < num_columns and col_index >=1:
            column = df_healthy_lh.columns[col_index]
            axes[i, j].boxplot([df_healthy_lh[column], df_mci_lh[column]], labels=['Healthy', 'MCI'])
            axes[i, j].set_title(column)

# Remove any empty subplots
for i in range(num_columns, num_rows * num_cols):
    fig.delaxes(axes.flatten()[i])

# Customize the layout
plt.tight_layout()
#plt.subplots_adjust(wspace=3.4, hspace=0.4)  # Adjust these values to increase or decrease spacing

# Show the plot
plt.show()
```

```python
# Analyse right hemisphere
# Create subplots for each column
num_columns = df_healthy_rh.shape[1]

# Create a figure with subplots
num_rows = 8
num_cols = 5
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 18))

# Iterate through columns
for i in range(num_rows):
    for j in range(num_cols):
        col_index = i * num_cols + j
        if col_index < num_columns and col_index >=1:
            column = df_healthy_rh.columns[col_index]
            axes[i, j].boxplot([df_healthy_rh[column], df_mci_rh[column]], labels=['Healthy', 'MCI'])
            axes[i, j].set_title(column)

# Remove any empty subplots
for i in range(num_columns, num_rows * num_cols):
    fig.delaxes(axes.flatten()[i])

# Customize the layout
plt.tight_layout()
#plt.subplots_adjust(wspace=3.4, hspace=0.4)  # Adjust these values to increase or decrease spacing

# Show the plot
plt.show()
```

```python
# Perform 2 sample t test on left hemisphere dataframes
t_test_results = []
for column in df_healthy_lh.columns[1:]:
    t_test_results.append(ttest_ind(df_healthy_lh[column], df_mci_lh[column]))
t_test_results_df = pd.DataFrame(t_test_results, columns=['t-statistic', 'p-value'])
print(f"t_test_results_df:\n {t_test_results_df}")
p_threshold = 0.05
for col_ind, p in enumerate(t_test_results_df['p-value']):
  if p < p_threshold:
    print(f"p-value for {df_healthy_lh.columns[col_ind+1]} is lower than 0.05: {p}")
```

```python
# Perform 2 sample t test on right hemisphere dataframes
t_test_results_rh = []
for column in df_healthy_rh.columns[1:]:
    t_test_results_rh.append(ttest_ind(df_healthy_rh[column], df_mci_rh[column]))
t_test_results_rh_df = pd.DataFrame(t_test_results_rh, columns=['t-statistic', 'p-value'])
print(f"t_test_results_rh_df:\n {t_test_results_rh_df}")
p_threshold = 0.05
for col_ind, p in enumerate(t_test_results_rh_df['p-value']):
  if p < p_threshold:
    print(f"p-value for {df_healthy_rh.columns[col_ind+1]} is lower than 0.05: {p}")
```
